```python
# функция выгружает в одну табличку num дней
def get_data_from_num_days(num):
  list_dir = os.listdir('transactions')
  list_dir.sort(reverse=True) # чтобы последних дней было больше
  step = len(list_dir) // num
  idx = np.arange(0, len(list_dir), step = step)
  file_names = np.array(list_dir)[idx]

  general_data = pd.read_csv(os.path.join('transactions', file_names[0]))
  for path in tqdm(file_names):
    general_data = general_data.append(pd.read_csv(os.path.join('transactions', path)),
                                       sort=False, ignore_index=True)

  return general_data
```

```python
data = get_data_from_num_days(20)
```
